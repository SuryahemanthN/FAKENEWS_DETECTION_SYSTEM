```python
true_df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/True.csv')
fake_df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Fake.csv')
```

```python
true_df.head()
```

```python
true_df['label']=1
fake_df['label']=0
```

```python
true_df.head()
```

```python
fake_df.head()
```

```python
news=pd.concat([true_df,fake_df],axis=0)
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news=news.drop(['title','subject','date'],axis=1)
news.head()
```

```python
news=news.sample(frac=1)
news.head()
```

```python
news.reset_index(inplace=True)
news.head()
```

```python
news=news.drop(['index'],axis=1)
news.head()
```

```python
def wordopt(text):
  text=text.lower()
  text=re.sub(r'<.*?>',' ',text)
  text=re.sub(r'http?://\S+|www\.\S+',' ',text)
  text=re.sub(r'[^\w\s]', ' ', text)
  text=re.sub(r'\d',' ',text)
  text=re.sub(r'\n',' ',text)
  return text
```

```python
news['text']=news['text'].apply(wordopt)
news['text']
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
xv_train=vectorizer.fit_transform(x_train)
xv_test=vectorizer.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc=DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()
```

```python
RFC.fit(xv_train,y_train)
```

```python
pred_rfc=RFC.predict(xv_test)
```

```python
RFC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
def output(n):
  if n==0:
    return 'It is a Fake News'
  elif n==1:
    return 'It is a Real News'
```

```python
def manual_testing(news):
  testing_news={"text":[news]}
  new_def_test=pd.DataFrame(testing_news)
  new_def_test["text"]=new_def_test["text"].apply(wordopt)
  new_x_test=new_def_test["text"]
  new_xv_test=vectorizer.transform(new_x_test)
  pred_lr=LR.predict(new_xv_test)
  pred_dtc=DTC.predict(new_xv_test)
  pred_rfc=RFC.predict(new_xv_test)
  return f"LR Prediction: {output(pred_lr[0])}, DTC Prediction: {output(pred_dtc[0])}, RFC Prediction: {output(pred_rfc[0])}"
```

```python
news_article=str(input("Enter an article "))
manual_testing(news_article)
```
